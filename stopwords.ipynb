## Mengenal apa itu stopwrods

Stopwords adalah kata-kata umum seperti “and”, “or”, dan sejenisnya, yang sering dianggap tidak penting dalam analisis teks, sehingga biasanya dihapus saat preprocessing agar fokus analisis lebih ke isi utama teks.

```python
#import packages
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
en_stopwords
```

```python
# menghilangkan stopwords dari sebuah kalimat
sentence = 'it was too far to go to the shop and he did not want her to walk'
clear_sentence = ' '.join(word for word in sentence.split() if word not in en_stopwords)
clear_sentence
```

```python
en_stopwords.remove('did')
en_stopwords.remove('not')
en_stopwords.append('go')
```
