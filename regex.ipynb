## Belajar tentang Regex

Regular Expression (regex) adalah sebuah metode atau alat yang digunakan untuk mendefinisikan pola (pattern) pada teks. Regex umumnya dimanfaatkan untuk mencari, mencocokkan, memvalidasi, atau memanipulasi data teks, seperti menghapus, mengganti, atau mengekstrak bagian tertentu sesuai dengan pola yang telah ditentukan.

```python
customer_reviews = ['sam was a great help to me in the store',
                   'the cashier was very rude to me, i think her name was elanor',
                   'amazing work from adeen!',
                   'sarah was able to help me find the items i needed quickly',
                   'lucy is such  great addition to the team',
                   'great services from sara she found me what i wanted']
```

```python
sarah_reviews = []
pattern_to_find = r"sarah?"

for i in customer_reviews:
    reviews = re.search(pattern_to_find, i)
    if reviews:
        sarah_reviews.append(i)
```

```python
sarah_reviews
```

```python
a_reviews = []
pattern_to_find = r"^a" #cari kalimat yang diawali dengan huruf 'a'

for string in customer_reviews:
    if (re.search(pattern_to_find, string)):
        a_reviews.append(string)
a_reviews
```

```python
y_reviews = []
pattern_to_find = r'y$' #cari kalimat yang akhirnya diakhiri dengan 'y'

for string in customer_reviews:
    if(re.search(pattern_to_find, string)):
        y_reviews.append(string)
y_reviews
```

```python
needwant_reviwes = []
pattern_to_find = r'(need|want)' # mencari kata need atau want

for string in customer_reviews:
    if(re.search(pattern_to_find, string)):
        needwant_reviwes.append(string)
needwant_reviwes
```

```python
no_punct_reviwes = []
pattern_to_find = r'[^\w\s]' # mencari kalimat yang tidak mengandung tanda baca

for string in customer_reviews:
    no_punct = re.sub(pattern_to_find, '', string)
    no_punct_reviwes.append(no_punct)

no_punct_reviwes
```
