## memecah kalimat menjadi token

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
#mencoba memecah sebuah kalimat menjadi beberapa kalimat
sentence_1 = "Her cat's name is Luna. Her dog's is max"
sent_tokenize(sentence_1)
```

```python
#mencoba memecah sebuah kalimat menjadi perkata
sentence_2 = "Her cat's name is luna"
word_tokenize(sentence_2)
```

```python
word_tokenize(sentence_1)
```
