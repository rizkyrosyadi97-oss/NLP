## Mengenal porterstemmer
berguna untuk merubah sebuah kata menjadi bentuk dasar, namun method ini memiliki kelemahan dimana hasil dari stemming kadang tdk sesuai dengan apa yang seharusnya

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
list_of_words = ['connecting', 'connectivity', 'connected', 'connects', 'connect']
```

```python
for t in list_of_words:
    print(t, ': ', ps.stem(t))
```

```python
list_of_word_2 = ['learning', 'learns', 'learned', 'learn', 'learner', 'learners']
```

```python
for t in list_of_word_2:
    print(t, ': ', ps.stem(t))
```

```python
list_of_word_3 = ['likes', 'worse', 'better']
```

```python
for t in list_of_word_3:
    print(t, ': ', ps.stem(t))
```

disana terlihat kelemahan dari stemming dimana worse dirubah menjadi wors artinya menjadi berbeda

## Lemmatizing
sedikit berbeda dari sebelumnya wordnetlemmatizer mereduce kata namun tetap mempertahankan maknanya, metode ini lebih baik namun juga memiliki kelemahan dimana akan lebih banyak memakan memory karena token"nya kebanyakan tidak akan berubah

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for t in list_of_words:
    print(t, ": ", lemmatizer.lemmatize(t))
```

```python
for t in list_of_word_2:
    print(t, ": ", lemmatizer.lemmatize(t))
```

```python
for t in list_of_word_3:
    print(t, ": ", lemmatizer.lemmatize(t))
```

terlihat kebanyakan token akan dipertahankan seperti semula untuk mempertahankan maknanya
